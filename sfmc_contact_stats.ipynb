```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

push_df = pd.read_csv('/Users/bo/Downloads/sfmc_csv/_mobilepush.csv')
mobile_df = pd.read_csv('/Users/bo/Downloads/sfmc_csv/_mobileaddress.csv')
subscribers_df = pd.read_csv('/Users/bo/Downloads/sfmc_csv/_subscribers.csv')

# Merge the dataframes using contact ID, with a fallback to subscriber key
merged_df = pd.merge(subscribers_df, push_df, on='contactid', how='outer') \
             .merge(mobile_df, on='contactid', how='outer', suffixes=('_push', '_mobile')) 
```

```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Convert the column of interest to sets
set1 = set(push_df['contactid'])
set2 = set(mobile_df['contactid'])
set3 = set(subscribers_df['contactid'])

# Plotting Venn diagram
venn3([set1, set2, set3],('Push', 'Mobile', 'Subscribers'))

# Display the plot
plt.show()
```

```python
merged_df = pd.merge(subscribers_df, push_df, on='contactid', how='outer') \
             .merge(mobile_df, on='contactid', how='outer', suffixes=('_push', '_mobile'))
```

```python

opt_in_count = len(subscribers_df[(subscribers_df['optin'] == 'true') & (subscribers_df['email'].notna())])

# Sample data
data = {
    'Contact Type': ['Reachable Email Contact', 'Overall Contact'],
    'Number': [opt_in_count, len(merged_df)]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Contact Type'], df['Number'], color=['blue', 'green'])

# Adding titles and labels
plt.title('Mail Contact Number vs Overall Contact Number')
plt.xlabel('Contact Type')
plt.ylabel('Number of Contacts')

# Displaying the plot
plt.show()

```

```python
opt_in_smscount = len(mobile_df[(mobile_df['optin'] == 'true') & (mobile_df['phonenumber'].notna())])

# Sample data
data = {
    'Contact Type': ['Reachable SMS Contact', 'Overall Contact'],
    'Number': [opt_in_smscount, len(merged_df)]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Contact Type'], df['Number'], color=['blue', 'green'])

# Adding titles and labels
plt.title('Mail Contact Number vs Overall Contact Number')
plt.xlabel('Contact Type')
plt.ylabel('Number of Contacts')

# Displaying the plot
plt.show()

```
